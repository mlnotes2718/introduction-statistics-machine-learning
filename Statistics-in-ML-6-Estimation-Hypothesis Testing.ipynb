Step 2: CI for a population mean

If you have:
- a sample mean $\bar{x}$
- a sample standard deviation $s$
- a sample size $n$

then a 95% confidence interval for the population mean is:

$$\bar{x} \pm t_{\alpha / 2, n - 1} \times \frac{s}{\sqrt{n}}$$

where $t_{\alpha / 2, n - 1}$ is the critical value from the t-distributions ($\approx 1.96$ if n is large)
